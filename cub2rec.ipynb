```python
%%bash
# This script converts the images in CUB-200-2011 to .rec files.
# Downloading the MXNet repo is necessary.
# git clone https://github.com/apache/incubator-mxnet.git
if [ ! -d "incubator-mxnet" ]; then
    git clone https://github.com/apache/incubator-mxnet.git
fi
if [ ! -d "dataset/io_data" ]; then
    mkdir dataset/io_data
fi
python3 incubator-mxnet/tools/im2rec.py --list --recursive --train-ratio .9 dataset/io_data/bird_data dataset/CUB_200_2011/images/ --exts .jpg
python3 incubator-mxnet/tools/im2rec.py --resize 224 --quality 80 --num-thread 16 dataset/io_data/bird_data dataset/CUB_200_2011/images/
```
